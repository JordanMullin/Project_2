```python
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo= False)

```

```python
Health_per_GDP= "Health_Spending_Per_of_GDP.csv"
Health_US_Spending="Health_Spending_US_Dollar_Per_Capita.csv"
```

```python
Health_per_GDP_df= pd.read_csv(Health_per_GDP)
Health_US_Spending_df= pd.read_csv(Health_US_Spending)
```

```python
Health_per_GDP_df.head(5)
```

```python
Merged_World_df = pd.merge(Health_per_GDP_df, Health_US_Spending_df, on="Country")
Merged_World_df.head()
```

```python
Merged_World_df= Merged_World_df[['Country','TIME_x','Value_x','Value_y']]
Merged_World_df.head(5)
```

```python
Merged_World_df=Merged_World_df.rename(columns={
    'Value_x':'Health_Spending_by_GDP',
    'Value_y':'Health_Spending_USD_by_Capita',
    'TIME_x':'Year'
})
```

```python
Covid_csv='Covid.csv'
Covid_df= pd.read_csv('Covid.csv')

```

```python
Covid_df = Covid_df.drop(columns=['Unnamed: 0'])
```

```python
Final_World_df= pd.merge(Merged_World_df, Covid_df, on="Country")
Final_World_df
```

```python
Final_World_df.loc[Final_World_df['Country'] == 'USA']
```

```python
Final_World_df = Final_World_df.rename(columns={
    "% of world": "percent_of_world",
    "Total Cases": "Total_Cases",
    "Total Deaths": "Total_Deaths",
    " Active Cases": "Active_Cases",
    "Total Cases/1M Population": "Cases_by_1M_Population",
    "Tests/1M Population": "Tests_by_1M_Population"  
})
```

```python
Final_World_df.to_csv('Final_World_df.csv', index=False)
```
