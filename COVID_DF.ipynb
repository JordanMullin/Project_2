```python
import pandas as pd
```

```python
covid_path = "Covid_project.csv"
population_path = "population_project.csv"

```

```python
covid_df = pd.read_csv(covid_path)
```

```python
covid_df = covid_df.rename(columns = {
    "5/1/2020":"Country", 
    "Unnamed: 1":"Total Cases", 
    "Unnamed: 2": "New Cases", 
    "Unnamed: 3": "Total Deaths", 
    "Unnamed: 4": "New Deaths", 
    "Unnamed: 5": "Total Recovered", 
    "Unnamed: 6":" Active Cases", 
    "Unnamed: 7":"Serious/Critical Cases", 
    "Unnamed: 8":"Total Cases/1M Population", 
    "Unnamed: 9":"Total Deaths/1M Population", 
    "Unnamed:10":"Total Tests", 
    "Unnamed: 11": "Tests/1M Population"})
```

```python
covid_df = covid_df.drop([0, 1])
```

```python
covid_df.head(30)
```

```python
pop_df = pd.read_csv(population_path)
```

```python
pop_df = pop_df.drop([0])
```

```python
complete_df = pd.merge(pop_df, covid_df, on="Country")
```

```python
complete_df = complete_df[["Country", "Population", "% of world", "Total Cases", "Total Deaths", " Active Cases", "Total Cases/1M Population", "Tests/1M Population" ]]
```
